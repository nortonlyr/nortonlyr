```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud
```

```python
string = 'Python, R, AWS, Spark, SQL, NoSQL, Kafka, Unittest, \
maplotlib, nltk, Airflow, Tableau, plotly, pandas, numpy, Java, BASH'

wc = WordCloud(background_color = 'white', 
              width = 1000, 
              height = 800, 
              ).generate(string)
wc.to_file('skills.png')
plt.imshow(wc)
plt.axis('off')
plt.show()
```
